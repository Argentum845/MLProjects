```python
# выбор столбцов для предсказания
selected = ["len", "wid", "loss", "st", "ns", "inj"]
data_prepared = data.loc[:, data.columns.intersection(selected)]
data_prepared.head()
```

```python
# просмотр, насколько часто встечается то или иное значение в столбце st (доля от общего числа)
data['st'].value_counts(normalize=True)
```

```python
# замена значений с малой долей вхождения на 1 общий ("прочее")
data_prepared['st'] = data_prepared['st'].replace(['RI', 'AK', 'DC', 'VI'],'Other')
data_prepared['st'].value_counts(normalize=True)
```

```python
# Факторизация (проверка ее работы) столбца st (замена категориальных данных на числовые)
data_prepared.st.factorize()
```

```python
# сама замена данных в столбце и просмотр, как это получилось 
data_prepared.st = data_prepared.st.factorize()[0]
data_prepared.head()
# data_prepared.describe()
```

```python
# удаление всех строк из датасета, где в столбце loss пустые значения (1 подход)
res = data_prepared.copy()
res.dropna(subset=['loss'], inplace=True)
res.info()
```

```python
# замена пустых значений на среднее арифметическое столбца (2 способ)
res = data_prepared.copy()
mean = res.loss.mean()
res.loss.fillna(mean, inplace=True)
res.info()
```
